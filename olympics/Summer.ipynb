# Summer Olympics Data Analysis Assignment

```python
# Importing libraries
import pandas as pd
```

```python
# Data loading
data = pd.read_csv("summer.csv")
data.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(data["City"].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold_data = data[data["Medal"] == "Gold"]
gold_data
sports = gold_data["Sport"].value_counts()
sports[:5]
```

```python
sports.plot(x=sports.index, y=sports.values, kind="bar", figsize=(15, 5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data["Medal"].unique()
```

```python
medals = data["Sport"].value_counts()
medals[:5]
```

```python
medals.plot(x=medals.index, y=medals.values, kind="bar", figsize=(15,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
players = data["Athlete"].value_counts()
players = players[:50]
players[:5]
```

```python
players.plot(x=players.index, y=players.values, kind="bar", figsize=(15,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
players_gold = gold_data["Athlete"].value_counts()
players_gold = players_gold[:50]
players_gold[:5]
```

```python
players_gold.plot(x=players_gold.index, y=players_gold.values, kind="bar", figsize=(15,5))
```
